```python
import numpy as np
import pandas as pd
import glob as g
```

```python
input = g.glob('in*')
print(input)
```

```python
def read_vars(file):
    with open(file) as f:
        rec = '!'
        while rec[0] == '!': rec = f.readline()  # skip initial comments
        flds = rec.split()
        nlev = int(flds[0])
        atm = { 'nlev':nlev } 
        rec = f.readline()
        while rec[0:4].lower() != '*end':  # repeat until end marker record
            if rec[0] == '!': continue 
            flds = rec.split()
            key = flds[0][1:].lower() # remove '*' and change to lower case
            prf = np.fromfile(f,sep=", ",count=nlev)
            atm[key] = prf
            rec = f.readline()
            if rec == '': break   # also exit if end-of-file without *END
    return atm
```

```python
atm = []
for x in input:
    atm.append(read_vars(x))
```

```python
print(atm[0])
print(len(atm))
```

```python
df_US = pd.DataFrame(atm[0], columns = ['hgt', 'pre', 'tem', 'co2'])
df_MLW = pd.DataFrame(atm[1], columns = ['hgt', 'pre', 'tem', 'co2'])
df_TROP = pd.DataFrame(atm[2], columns = ['hgt', 'pre', 'tem', 'co2'])
df_SAW = pd.DataFrame(atm[3], columns = ['hgt', 'pre', 'tem', 'co2'])
df_MLS = pd.DataFrame(atm[4], columns = ['hgt', 'pre', 'tem', 'co2'])
df_SAS = pd.DataFrame(atm[5], columns = ['hgt', 'pre', 'tem', 'co2'])
```

```python
df_US.to_csv('USstd_CO2.csv', sep=' ', header=False)
df_MLS.to_csv('MLS_CO2.csv', sep=' ', header=False)
df_MLW.to_csv('MLW_CO2.csv', sep=' ', header=False)
df_SAW.to_csv('SAW_CO2.csv', sep=' ', header=False)
df_SAS.to_csv('SAS_CO2.csv', sep=' ', header=False)
df_TROP.to_csv('TROP_CO2.csv', sep=' ', header=False)
```
